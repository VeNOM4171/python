## in table we want to do combine selection of two column name 
## like: to display the countries where there are less than 100 speaker.

```python
data.rename(columns={'Number of speakers' : 'no_of_speaker'}, inplace=True)
data
```

```python
data[(data.Countries == 'India') & (data.no_of_speaker < 100)]
```

##  GroupBy-- sort no_of_speaker

```python
data.sort_values('no_of_speaker',ascending=True)
```
